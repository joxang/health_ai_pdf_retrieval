```python
## pulls first href (which is always the direct access link) from the google scholar search page
## searches for each doi = 1 result only

href_list = []
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/36.0.1941.0 Safari/537.36'}

for doi in tqdm(doi_list):
    
    ## >60s seems to prevent being locked out
    sleep(randint(61,63))
    
    ## returns response  code
    page = requests.get('https://scholar.google.com/scholar?q=' + doi, headers=headers)
    print(page)
    
    ## content to tree
    tree = html.fromstring(page.content)    
    hrefs = tree.xpath('//div[@class="gs_or_ggsm"]/a')
    
    ## try to print first link, unless no link
    try:
        print(hrefs[0].attrib['href'])
        href_list.append(hrefs[0].attrib['href'])
    
    except:
        print('nolink')
        href_list.append('nolink')
```
