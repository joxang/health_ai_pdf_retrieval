# webdriver to download pdfs from new lists

```python
## compatible html links to pdf links (most have a javascript redirect)

pmid_list = missing_df['pmid'].tolist()

scrape_pdf_list = []
pmid_pdf_list = []

for pmid, link in zip(pmid_list, scrape_list):

    if "mdpi.com" in link:
        pmid_pdf_list.append(pmid)
        newlink = link + "/pdf"
        #print(newlink)
        scrape_pdf_list.append(newlink)

    elif "wiley.com" in link:
        pmid_pdf_list.append(pmid)        
        index = link.find("wiley.com/doi/")
        newlink = link[:index+14] + "pdfdirect/" + link[index+14:]
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    elif "hindawi.com" in link:
        pmid_pdf_list.append(pmid)        
        index = link.find("hindawi.com/")
        newlink = "https://" + link[12:index] + "downloads." + link[index:-1] +".pdf"
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    elif "sagepub.com" in link:
        pmid_pdf_list.append(pmid)        
        index = link.find("sagepub.com/doi/")
        newlink = link[:index+16] + "pdf/" + link[index+16:]
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    elif "springer.com" in link:
        pmid_pdf_list.append(pmid)        
        index = link.find("springer.com/")
        newlink = link[:index+13] + "content/pdf" + link[index+20:] +".pdf"
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    elif "frontiersin.org" in link:
        pmid_pdf_list.append(pmid)        
        newlink = link[:-4] + "pdf"
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    elif "tandfonline" in link:
        pmid_pdf_list.append(pmid)        
        index = link.find("full/")
        newlink = link[:index] + "pdf/" + link[index+5:]
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    elif "nature.com" in link:
        pmid_pdf_list.append(pmid)        
        newlink = link + ".pdf"
        #print(newlink)
        scrape_pdf_list.append(newlink)
        
    else:
        pass
        #scrape_pdf_list.append(np.nan)

print(len(scrape_pdf_list))
print(len(pmid_pdf_list))
```

```python
## resets user agent each loop
from user_agent import generate_user_agent, generate_navigator

## downloads pdf into folder named by pmid for each paper
for pmid, url in tqdm(zip(pmid_pdf_list, scrape_pdf_list)):
    
    ## reset options
    options = webdriver.ChromeOptions() 

    userAgent = generate_user_agent()

    options.add_experimental_option("excludeSwitches", ["enable-logging"])
    options.add_argument(f'user-agent={userAgent}')
    profile = {
        'download.prompt_for_download': False,
        'download.directory_upgrade': True,
        'download.default_directory' : 'C:\\pdf\\' + str(pmid),
        'plugins.always_open_pdf_externally': True
    }
    options.add_experimental_option('prefs', profile)
    
    ## initialise driver
    driver = webdriver.Chrome(options=options)
    
    ## modify navigator webdriver flag
    driver.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
      "source": """
        Object.defineProperty(navigator, 'webdriver', {
          get: () => undefined
        })
      """
    })

    ## get url and wait for redirect
    driver.get(url)

    time.sleep(8) # >6s required to allow redirect and download of larger pdfs

    driver.quit()
```

# webdriver to download ieee files

```python
## list of ieee jsp pages (which have button to pdf)

pmid_list = missing_df['pmid'].tolist()

scrape_pdf_list = []
pmid_pdf_list = []

for pmid, link in zip(pmid_list, scrape_list):

    if "ieee.org" in link:
        pmid_pdf_list.append(pmid)
        index = link.find("document/")
        newlink = link[:index] + "/stamp/stamp.jsp?tp=&arnumber=" + link[index+9:]
        #print(newlink)
        scrape_pdf_list.append(newlink)
    
    else:
        pass
    
print(len(pmid_pdf_list))
print(len(scrape_pdf_list))
```

```python
## resets user agent each loop
from user_agent import generate_user_agent, generate_navigator

## downloads pdf into folder named by pmid for each paper
for pmid, url in tqdm(zip(pmid_pdf_list, scrape_pdf_list)):
    
    ## reset options
    options = webdriver.ChromeOptions() 

    userAgent = generate_user_agent()

    options.add_experimental_option("excludeSwitches", ["enable-logging"])
    options.add_argument(f'user-agent={userAgent}')
    profile = {
        'download.prompt_for_download': False,
        'download.directory_upgrade': True,
        'download.default_directory' : 'C:\\pdf\\' + str(pmid),
        'plugins.always_open_pdf_externally': True
    }
    options.add_experimental_option('prefs', profile)
    
    ## initialise driver
    driver = webdriver.Chrome(options=options)
    
    ## modify navigator webdriver flag
    driver.execute_cdp_cmd("Page.addScriptToEvaluateOnNewDocument", {
      "source": """
        Object.defineProperty(navigator, 'webdriver', {
          get: () => undefined
        })
      """
    })

    ## get url and wait for redirect
    
    try:
        driver.get(url)

        time.sleep(5) # load frames

        ## switch to iframe (only single frame on these pages) and click button
        driver.switch_to.frame(1)
    
        button = driver.find_element(by=By.XPATH, value='//button[@id="open-button"][.="Open"]')
        button.click()
        time.sleep(10) # enough time to download pdf

        driver.quit()
        
    except:
        driver.quit()
        continue
        
```
